## **Instructions:**  

## **1. Generate and Visualize Data**  

### **Task:**  
- Create a dataset containing **150 samples** with a **linear relationship**:  
  $$ y = 4 + 3x + \text{noise} $$
- Use `numpy` to generate random values for **X**.
- Plot the dataset using `matplotlib` to visualize the data points.


```python
import numpy as np

np.random.seed(42)

X = 2 * np.random.rand(150,1)  
noise = np.random.randn(150,1) 

y = 4 + 3 * X + noise
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Data that was Generated")
plt.show()
```

## **2. Compute Optimal Parameters using the Normal Equation**  

### **Task:**  
- Add a **bias term (column of ones)** to the feature matrix **X**.  
- Compute the optimal **theta parameters** using the **Normal Equation**:  
  $$ \theta = (X^TX)^{-1}X^Ty $$
- Print the computed **intercept** and **coefficient** values.



```python
X_2 = np.c_[np.ones((150,1)), X]

theta = np.linalg.inv(X_2.T @ X_2) @ X_2.T @ y

print("Intercept values:" , theta[0])
print("Coefficient values:" , theta[1])
```

## **3. Visualize the Fitted Line (Normal Equation)**  

### **Task:**  
- Use the computed **theta** values from the **Normal Equation** to plot the best-fit line on the dataset.
- Overlay the **original data points** with the **fitted regression line**.



```python
y_pred = X_2 @ theta 

plt.scatter(X,y)

plt.plot(X, y_pred, color = "red")

plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression fit (Normal Equation)")
plt.show()
```

## **4. Train a Linear Model using Scikit-Learn's `LinearRegression`**  

### **Task:**  
- Train a linear regression model using **`LinearRegression`** from `sklearn`.  
- Print the **intercept** and **coefficient** values.  


```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X,y)

print("Intercept values: ", model.intercept_)
print("Coefficient values: ", model.coef_)
```

```python
y_pred_m = model.predict(X)

plt.scatter(X, y)

plt.plot(X, y_pred, color = "red")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression Fit by Scikit-learn")
plt.show()
```
